```python
import konlpy
from konlpy.tag import *
import pandas as pd
import re
```

```python
#한나눔 - KAIST 개발 --> 추천
hannanum = Hannanum()
#꼬꼬마 - 서울대 개발
kkma = Kkma()
```

```python
data = pd.read_csv("./top_topic_news_20230817.csv", encoding='utf-8')
```

```python
new_list = []

for k in data['title']:
    #html 태그 삭제
    k2 = re.sub('<.+?>', '', str(k), 0).strip()
    #특수기호 삭제
    k3 = re.sub('[^ A-Za-z0-9가-힣]+', ' ', str(k2))
    n = hannanum.nouns(k3)
    print (n)
    new_list.append(n)

data['title_hann'] = new_list
```

```python
content_list = []

for k in data['content']:
    #html 태그 삭제
    k2 = re.sub('<.+?>', '', str(k), 0).strip()
    #특수기호 삭제
    k3 = re.sub('[^ A-Za-z0-9가-힣]+', ' ', str(k2))
    n1 = hannanum.nouns(k3)
    print (n1)
    content_list.append(n1)

data['content_hann'] = content_list
```

```python
new_df = data['title_hann','content_hann']
```

```python
new_df.to_csv('top_tp.csv',index=False, encoding="utf-8-sig")
```
