```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd
import re 
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options

chrome_options = Options()
chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)
```

```python
def click_first_post(driver):
    try:
        first_post = driver.find_element(By.CSS_SELECTOR, 'article > div:nth-child(1) a')
        first_post.click()
        time.sleep(3)
    except Exception as e:
        print("Error:", e)
        
def get_content(driver):

    # 1. 현재 게시글 html 정보 가져오기
    html = driver.page_source
    soup = BeautifulSoup(html, 'lxml')
    try:
        content = soup.select('div._a9zs')[0].text
    except:
        content = ''
    data = [content]
    return data

def next_page(driver):
    next_page = driver.find_element(By.CSS_SELECTOR,'div._aaqg._aaqh')
    next_page.click()
    time.sleep(3)
```

```python
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

driver.get('https://www.instagram.com/accounts/login/')
driver.maximize_window()
time.sleep(4)

username='woori_won1'
password='woori1234!'
user_id= driver.find_element(By.CSS_SELECTOR, '#loginForm > div > div:nth-child(1) > div > label > input')
user_id.send_keys(username)
time.sleep(3)

user_id= driver.find_element(By.CSS_SELECTOR, '#loginForm > div > div:nth-child(2) > div > label > input')
user_id.send_keys(password)
time.sleep(3)

driver.find_element(By.CSS_SELECTOR, '#loginForm > div > div:nth-child(3)').click()
time.sleep(3)
```

```python
driver.get('https://www.instagram.com/kbkookminbank/')
time.sleep(4)

click_first_post(driver)
results=[]

target=1300
for i in range(target):
    try:
        data=get_content(driver)
        results.append(data)
        next_page(driver)
    except:
        time.sleep(2)
        next_page(driver)
    time.sleep(2)
    print("현재까지 가져온 데이터 개수:", len(results))
    
driver.quit()
print(results[:2])
```
